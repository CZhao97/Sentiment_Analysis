```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Bernanke_A_Article.csv')
```

```python
df.head()
```

```python
df = df[df['sentiment_head']!=0]
```

```python
df.head()
```

```python
df.groupby(['Numeric Date']).sentiment_head.agg(lambda x: sum(x)/len(x)).plot.line(x='Numeric Date', y='ahkz_sentiment_average', title='sentiment_head Over Days', rot=15)
plt.tight_layout()
plt.savefig('../plots/article_level_head/Bernanke Group A Article Daily Count.PNG')
```

```python
df['Numeric Date'] = pd.to_datetime(df['Numeric Date'], format='%Y/%m/%d').apply(lambda x: x.strftime('%Y-%m'))
df = df.sort_values('Numeric Date')
```

```python
df.head()
```

```python
df.groupby(['Numeric Date']).sentiment_head.agg(lambda x: sum(x)/len(x)).plot.line(x='Numeric Date', y='sentiment_head', title='sentiment_head Over Months', rot=15)
plt.tight_layout()
plt.savefig('../plots/article_level_head/Bernanke Group A Article Monthly Count.PNG')
```
